```python
# helper funcs
def ask(prompt: str, condition: callable = None) -> str:
    print('( ожидаю ввод текста )')
    prepending_str = ''
    while True:
        result = input(f'{prepending_str}{prompt}\n >> ')
        if condition is None or condition(result):
            print(f'> {result}')
            return result
        else:
            prepending_str = 'неверное значение. '

def ask_yn() -> bool:
    return ask('да/нет', lambda x: x.lower() in YN) == 'да'

def ask_yn_force() -> None:
    ask('да/нет', lambda x: x == 'да')

def space(lines: int = 1) -> None:
    print('\n' * (lines - 1))


# class to replace `random` lib
class Random:
    def __init__(self, seed: str):
        self.numhash = self.__numhash(seed)

    def __numhash(self, string: str) -> list[int]:
        num = abs(hash(string))
        num_str = str(num) + str(len(string))
        digit_list = [int(digit) for digit in num_str]
        return digit_list
    
    def __relative_index(self, array, offset):
        idx = round(offset * (len(array) - 1))
        return array[idx]

    def get(self, offset: float) -> int:
        return self.__relative_index(self.numhash, offset)


# consts
PROGRAMS = ['городское планирование', 'клеточная и молекулярная биотехнология', 'когнитивная нейробиология', 'ассириология: языки и история древней месопотамии', 'востоковедение', 'юго-восточная азия: языки, история, культуры', 'япония: язык, история, литература, культура', 'дизайн', 'мода', 'дизайн и разработка информационных продуктов', 'информатика и вычислительная техника', 'компьютерные науки и технологии', 'компьютерные технологии, системы и сети', 'программная инженерия', 'программная инженерия (очно-заочное обучение)', 'программные системы и автоматизация процессов разработки', 'разработка информационных систем для бизнеса', 'технологии искусственного и дополненного интеллекта', 'информационная безопасность', 'компьютерная безопасность', 'история искусств', 'современное искусство', 'античность', 'история', 'культурология', 'вычислительные социальные науки', 'компьютерные науки и анализ данных', 'математика', 'прикладная математика', 'прикладная математика и информатика', 'прикладной анализ данных', 'прикладной анализ данных и искусственный интеллект', 'совместный бакалавриат ниу вшэ и цпм', 'география глобальных изменений и геоинформационные технологии', 'международная программа «международные отношения и глобальные исследования»', 'международные отношения', 'политология', 'политология и мировая политика', 'программа двух дипломов ниу вшэ и университета кёнхи «экономика и политика в азии»', 'психология', 'социология', 'социология и социальная информатика', 'глобальные цифровые коммуникации', 'журналистика', 'кинопроизводство', 'медиакоммуникации', 'реклама и связи с общественностью', 'стратегия и продюсирование в коммуникациях', 'управление в креативных индустриях', 'актер', 'физика', 'философия', 'химия', 'аналитика в экономике', 'бизнес-информатика', 'государственное и муниципальное управление', 'маркетинг и рыночная аналитика', 'международная программа по экономике и финансам', 'международный бакалавриат по бизнесу и экономике', 'международный бизнес', 'мировая экономика', 'программа двух дипломов ниу вшэ и рут «экономика и инженерия транспортных систем»', 'совместная программа по экономике ниу вшэ и рэш', 'управление бизнесом', 'управление и аналитика в государственном секторе', 'управление цепями поставок и бизнес-аналитика', 'управление цифровым продуктом', 'цифровой маркетинг', 'экономика', 'экономика и анализ данных', 'экономика и бизнес (очно-заочное обучение)', 'экономика и статистика', 'экономический анализ', 'инфокоммуникационные технологии и системы связи', 'юриспруденция', 'юриспруденция: цифровой юрист', 'иностранные языки и межкультурная бизнес-коммуникация', 'иностранные языки и межкультурная коммуникация', 'иностранные языки и межкультурная коммуникация в бизнесе', 'филология', 'фундаментальная и компьютерная лингвистика', 'фундаментальная и прикладная лингвистика']
YN = ['да', 'нет']
EXAM_QUESTIONS = {
    'Как называется созвучие концов строк в стихотворении (опустелом — телом; тишины — стены)?': 'рифма',
    'Вероятность того, что мотор холодильника прослужит более 1 года, равна 0,8, а вероятность того, что он прослужит более 2 лет, равна 0,6. Какова вероятность того, что мотор прослужит более 1 года, но не более 2 лет?': '0.2',
    'Расставьте знаки препинания: укажите все цифры, на месте которых в предложении должны стоять запятые.\nИлюша был впечатлительным ребёнком (1) и очень любил слушать сказки (2) которые рассказывала ему няня (3) потому что в них было много чудесного (4) а (5) добро всегда побеждало зло.': '234'
}
COMPANIES = ['ООО "ЯНДЕКС"', 'ООО "В КОНТАКТЕ"', 'ООО "АВИТО"', 'ООО "ВЫСШАЯ ШКОЛА СЕРФ КОФЕ ХИЛЛС"']


# vars
state = {
    'user_name': None,
    'program': None,
    'is_commercial': None,
    'primary_way': None,
    'is_accepted': False,
}
required_documents = {
    'СНИЛС': '.jpg', 
    'Паспорт (1 стр.)': '.png',
    'Паспорт (стр. с пропиской)': '.webp',
    'Аттестат': '.pdf',
    'Согласие об обработке ПД': '.docx',
    'Фото 3 на 4': '.tiff',
    'Размер обуви': '.txt',
    'Диплом "Русский Медвежонок"': '.pdf',
    'Фото кредитной карты с двух сторон': '.heif',
}
application_troubles = ['file_size', 'file_name', 'server_error', 'server_maintainance', 'server_flood', 'server_hacked']


# main code
print('привет! добро пожаловать в симулятор поступления в ниу вшэ.\nэто очень сложный и непонятный путь, и интересно не будет.\n\nдля начала — давай знакомиться.')
state['user_name'] = ask('как тебя зовут?', lambda x: (len(x) > 1))

rnd = Random(state['user_name'])

space()
print(f"{state['user_name'].capitalize()}, нам надо выбрать ОП. это такая штука, на которую придется потратить ближайшие 4 года.\nвыбрать подходящую можно на сайте вышки: https://www.hse.ru/education/\n\nпожалуйста, напиши, какую ОП ты выбрал(а).")
state['program'] = ask('на какую оп поступаем? например, "дизайн"', lambda x: x.lower() in PROGRAMS)

space()
print(f'интересный выбор. год обучения на этой программе стоит {rnd.get(0.1) + 1 * 345678} рублей.\nесть ли у тебя такие деньги?')
state['is_commercial'] = ask_yn()

space()
if state['is_commercial']:
    exam_required_score = 180
    print('гудфою! тогда для поступления нужно набрать всего 180 баллов егэ в сумме.')
    state['primary_way'] = 'exam'
else:
    exam_required_score = 300 - rnd.get(0.3)
    print(f'неудивительно. тогда придется поступать бесплатно. давай оценим твои шансы на это.\n\nпо егэ для бюджета придется набрать {exam_required_score} баллов. Осилим?')
    if ask_yn():
        space()
        print('хорошо, будем пытаться поступать по егэ.')
        state['primary_way'] = 'exam'
    else:
        space()
        print('ну и правильно, лучше даже не пытаться.\nтогда может социальный лифт — ты случайно не из неблагополучной семьи?')
        if ask_yn():
            exam_required_score = int(exam_required_score * 0.75)
            space()
            print(f'это неплохо повышает твои шансы! будешь в отдельном списке на поступление, где проходной балл — {exam_required_score}. уже лучше!')
            state['primary_way'] = 'exam'
        else:
            space()
            print('в данном контексте даже хочется сказать "жаль". ну ладно.\nследующий вариант — целевая квота. попробуем найти компанию, которая захочет оплатить обучение за тебя?')
            if ask_yn():
                space()
                print(f'здорово! давай пробовать')
                state['primary_way'] = 'quota'
            else:
                space()
                print('что-ж, наша последняя опция — олимпиады. других вариантов нет.')
                state['primary_way'] = 'olymp'                 

space()
print('приступаем к непосредственному поступлению. это симулятор, поэтому если не выйдет — всегда можно попробовать снова. пока не надоест. ну, или пока не поступишь. готов?')
ask_yn_force()

while True:
    space()
    if state['primary_way'] == 'exam':
        print('сдаем егэ. не нервничаем, вспоминаем к чему готовились.')
        score = 0
        for k, v in EXAM_QUESTIONS.items():
            space()
            print(k)
            ans = ask('твой ответ')
            if ans == v:
                score += 100
        
        space()
        score += rnd.get(0.4)

        if score >= exam_required_score:
            print(f'ты набрал {score} баллов при пороге в {exam_required_score}. ура!!! поздравляю!!!')
            state['is_accepted'] = True
        else:
            print(f'ты набрал {score} баллов, а надо было {exam_required_score}. увы... поступление на этом закончилось.')
            state['is_accepted'] = False

    elif state['primary_way'] == 'quota':
        print('есть несколько компаний. заинтересуешь ли ты их — вопрос. но выбрать можно только одну. выбирай!')
        for index, i in enumerate(COMPANIES):
            print(f'{index + 1}. {i}')
        choice = ask('введи число', lambda x: x.isdigit() and int(x) > 0 and int(x) <= len(COMPANIES))

        space()
        success = (int(choice) + rnd.get(0.4)) % 2 == 0
        if success:
            print('эта компания оплатит твое обучение! ура!!! поздравляю!!!')
            state['is_accepted'] = True
        else:
            print('этой компании не очень интересно. увы... поступление на этом закончилось.')
            state['is_accepted'] = False

    elif state['primary_way'] == 'olymp' or True:
        print('егэ, как известно, лотерея, но олимпиады — лотерея в кубе. поэтому тут в целях симуляции даже не будет реальных вопросов — я предлагаю ткнуть пальцем в небо и назвать число от 1 до 256.')
        choice = ask('введи число', lambda x: x.isdigit() and int(x) > 0 and int(x) <= 256)
        
        space()
        success = (int(choice) + rnd.get(0.4)) % 3 == 0
        if success:
            print('ты в числе победителей! ура!!! поздравляю!!!')
            state['is_accepted'] = True
        else:
            print(f'упс, тебя обошло {125 * (rnd.get(0.5) + int(choice))} других учеников. увы... поступление на этом закончилось.')
            state['is_accepted'] = False


    if state['is_accepted']:
        break
    else:
        space()
        print('пробуем снова?')
        if not ask_yn(): break

space()
print('пришла пора подавать документы. все, что было до этого, было гораздо легче, так что удачи. если что-то пойдет не так — придется начинать сначала.\nвот список документов:')

space()
for k, v in required_documents.items():
    print(f'{k} — слать в формате {v}')

space()
print('приступаем?')
ask_yn_force()

attempt = 0
while True:
    for k, v in required_documents.items():
        space()
        print(f'приложи {k}')
        file = ask('напиши название файла в нужном формате')
        attempt += 1
        if not file.endswith(v):
            print(f'неверный формат файла! нужен был {v}. заново.')
            break
        if attempt % 4 == 0 and len(application_troubles) > 0:
            trouble = application_troubles.pop(0)
            if trouble == 'file_size':
                print('ой! размер слишком большой. заново.')
            elif trouble == 'file_name':
                print('ой! имя не подходит. заново.')
            elif trouble == 'server_error':
                print('ой! серверная ошибка. заново.')
            elif trouble == 'server_maintainance':
                print('ой! на сервере работы. заново.')
            elif trouble == 'server_flood':
                print('ой! сервер затопило... блблблблл...🐟🐟🐟... заново.')
            elif trouble == 'server_hacked':
                print('ой! сервер взломали... 起來！不願做奴隸的人們！заново.')
            else:
                print('ой! неизвестная ошибка. заново.')
            break
    else:
        break

space(2)
print('ура! твоя заявка готова. отправляем??')
ask_yn_force()

space()
print('500 Internal Server Error')

story = 'ты пытался поступить в ниу вшэ.\n'

for k, v in state.items():
    if k == 'user_name':
        story += f'тебя звали {v}\n'
    elif k == 'program':
        story += f'ты хотели на оп {v}\n'
    elif k == 'is_commercial':
        story += f"вариант — {'коммерция' if v else 'бюджет'}\n"
    elif k == 'is_accepted':
        story += f"тебя {'взяли' if v else 'не взяли'}\n"

with open('story.txt', 'w') as file:
    file.write(story)
```
