```python
# helper funcs
def ask(prompt: str, condition: callable = None) -> str:
    print('( –æ–∂–∏–¥–∞—é –≤–≤–æ–¥ —Ç–µ–∫—Å—Ç–∞ )')
    prepending_str = ''
    while True:
        result = input(f'{prepending_str}{prompt}\n >> ')
        if condition is None or condition(result):
            print(f'> {result}')
            return result
        else:
            prepending_str = '–Ω–µ–≤–µ—Ä–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ. '

def ask_yn() -> bool:
    return ask('–¥–∞/–Ω–µ—Ç', lambda x: x.lower() in YN) == '–¥–∞'

def ask_yn_force() -> None:
    ask('–¥–∞/–Ω–µ—Ç', lambda x: x == '–¥–∞')

def space(lines: int = 1) -> None:
    print('\n' * (lines - 1))


# class to replace `random` lib
class Random:
    def __init__(self, seed: str):
        self.numhash = self.__numhash(seed)

    def __numhash(self, string: str) -> list[int]:
        num = abs(hash(string))
        num_str = str(num) + str(len(string))
        digit_list = [int(digit) for digit in num_str]
        return digit_list
    
    def __relative_index(self, array, offset):
        idx = round(offset * (len(array) - 1))
        return array[idx]

    def get(self, offset: float) -> int:
        return self.__relative_index(self.numhash, offset)


# consts
PROGRAMS = ['–≥–æ—Ä–æ–¥—Å–∫–æ–µ –ø–ª–∞–Ω–∏—Ä–æ–≤–∞–Ω–∏–µ', '–∫–ª–µ—Ç–æ—á–Ω–∞—è –∏ –º–æ–ª–µ–∫—É–ª—è—Ä–Ω–∞—è –±–∏–æ—Ç–µ—Ö–Ω–æ–ª–æ–≥–∏—è', '–∫–æ–≥–Ω–∏—Ç–∏–≤–Ω–∞—è –Ω–µ–π—Ä–æ–±–∏–æ–ª–æ–≥–∏—è', '–∞—Å—Å–∏—Ä–∏–æ–ª–æ–≥–∏—è: —è–∑—ã–∫–∏ –∏ –∏—Å—Ç–æ—Ä–∏—è –¥—Ä–µ–≤–Ω–µ–π –º–µ—Å–æ–ø–æ—Ç–∞–º–∏–∏', '–≤–æ—Å—Ç–æ–∫–æ–≤–µ–¥–µ–Ω–∏–µ', '—é–≥–æ-–≤–æ—Å—Ç–æ—á–Ω–∞—è –∞–∑–∏—è: —è–∑—ã–∫–∏, –∏—Å—Ç–æ—Ä–∏—è, –∫—É–ª—å—Ç—É—Ä—ã', '—è–ø–æ–Ω–∏—è: —è–∑—ã–∫, –∏—Å—Ç–æ—Ä–∏—è, –ª–∏—Ç–µ—Ä–∞—Ç—É—Ä–∞, –∫—É–ª—å—Ç—É—Ä–∞', '–¥–∏–∑–∞–π–Ω', '–º–æ–¥–∞', '–¥–∏–∑–∞–π–Ω –∏ —Ä–∞–∑—Ä–∞–±–æ—Ç–∫–∞ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–æ–Ω–Ω—ã—Ö –ø—Ä–æ–¥—É–∫—Ç–æ–≤', '–∏–Ω—Ñ–æ—Ä–º–∞—Ç–∏–∫–∞ –∏ –≤—ã—á–∏—Å–ª–∏—Ç–µ–ª—å–Ω–∞—è —Ç–µ—Ö–Ω–∏–∫–∞', '–∫–æ–º–ø—å—é—Ç–µ—Ä–Ω—ã–µ –Ω–∞—É–∫–∏ –∏ —Ç–µ—Ö–Ω–æ–ª–æ–≥–∏–∏', '–∫–æ–º–ø—å—é—Ç–µ—Ä–Ω—ã–µ —Ç–µ—Ö–Ω–æ–ª–æ–≥–∏–∏, —Å–∏—Å—Ç–µ–º—ã –∏ —Å–µ—Ç–∏', '–ø—Ä–æ–≥—Ä–∞–º–º–Ω–∞—è –∏–Ω–∂–µ–Ω–µ—Ä–∏—è', '–ø—Ä–æ–≥—Ä–∞–º–º–Ω–∞—è –∏–Ω–∂–µ–Ω–µ—Ä–∏—è (–æ—á–Ω–æ-–∑–∞–æ—á–Ω–æ–µ –æ–±—É—á–µ–Ω–∏–µ)', '–ø—Ä–æ–≥—Ä–∞–º–º–Ω—ã–µ —Å–∏—Å—Ç–µ–º—ã –∏ –∞–≤—Ç–æ–º–∞—Ç–∏–∑–∞—Ü–∏—è –ø—Ä–æ—Ü–µ—Å—Å–æ–≤ —Ä–∞–∑—Ä–∞–±–æ—Ç–∫–∏', '—Ä–∞–∑—Ä–∞–±–æ—Ç–∫–∞ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–æ–Ω–Ω—ã—Ö —Å–∏—Å—Ç–µ–º –¥–ª—è –±–∏–∑–Ω–µ—Å–∞', '—Ç–µ—Ö–Ω–æ–ª–æ–≥–∏–∏ –∏—Å–∫—É—Å—Å—Ç–≤–µ–Ω–Ω–æ–≥–æ –∏ –¥–æ–ø–æ–ª–Ω–µ–Ω–Ω–æ–≥–æ –∏–Ω—Ç–µ–ª–ª–µ–∫—Ç–∞', '–∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–æ–Ω–Ω–∞—è –±–µ–∑–æ–ø–∞—Å–Ω–æ—Å—Ç—å', '–∫–æ–º–ø—å—é—Ç–µ—Ä–Ω–∞—è –±–µ–∑–æ–ø–∞—Å–Ω–æ—Å—Ç—å', '–∏—Å—Ç–æ—Ä–∏—è –∏—Å–∫—É—Å—Å—Ç–≤', '—Å–æ–≤—Ä–µ–º–µ–Ω–Ω–æ–µ –∏—Å–∫—É—Å—Å—Ç–≤–æ', '–∞–Ω—Ç–∏—á–Ω–æ—Å—Ç—å', '–∏—Å—Ç–æ—Ä–∏—è', '–∫—É–ª—å—Ç—É—Ä–æ–ª–æ–≥–∏—è', '–≤—ã—á–∏—Å–ª–∏—Ç–µ–ª—å–Ω—ã–µ —Å–æ—Ü–∏–∞–ª—å–Ω—ã–µ –Ω–∞—É–∫–∏', '–∫–æ–º–ø—å—é—Ç–µ—Ä–Ω—ã–µ –Ω–∞—É–∫–∏ –∏ –∞–Ω–∞–ª–∏–∑ –¥–∞–Ω–Ω—ã—Ö', '–º–∞—Ç–µ–º–∞—Ç–∏–∫–∞', '–ø—Ä–∏–∫–ª–∞–¥–Ω–∞—è –º–∞—Ç–µ–º–∞—Ç–∏–∫–∞', '–ø—Ä–∏–∫–ª–∞–¥–Ω–∞—è –º–∞—Ç–µ–º–∞—Ç–∏–∫–∞ –∏ –∏–Ω—Ñ–æ—Ä–º–∞—Ç–∏–∫–∞', '–ø—Ä–∏–∫–ª–∞–¥–Ω–æ–π –∞–Ω–∞–ª–∏–∑ –¥–∞–Ω–Ω—ã—Ö', '–ø—Ä–∏–∫–ª–∞–¥–Ω–æ–π –∞–Ω–∞–ª–∏–∑ –¥–∞–Ω–Ω—ã—Ö –∏ –∏—Å–∫—É—Å—Å—Ç–≤–µ–Ω–Ω—ã–π –∏–Ω—Ç–µ–ª–ª–µ–∫—Ç', '—Å–æ–≤–º–µ—Å—Ç–Ω—ã–π –±–∞–∫–∞–ª–∞–≤—Ä–∏–∞—Ç –Ω–∏—É –≤—à—ç –∏ —Ü–ø–º', '–≥–µ–æ–≥—Ä–∞—Ñ–∏—è –≥–ª–æ–±–∞–ª—å–Ω—ã—Ö –∏–∑–º–µ–Ω–µ–Ω–∏–π –∏ –≥–µ–æ–∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–æ–Ω–Ω—ã–µ —Ç–µ—Ö–Ω–æ–ª–æ–≥–∏–∏', '–º–µ–∂–¥—É–Ω–∞—Ä–æ–¥–Ω–∞—è –ø—Ä–æ–≥—Ä–∞–º–º–∞ ¬´–º–µ–∂–¥—É–Ω–∞—Ä–æ–¥–Ω—ã–µ –æ—Ç–Ω–æ—à–µ–Ω–∏—è –∏ –≥–ª–æ–±–∞–ª—å–Ω—ã–µ –∏—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏—è¬ª', '–º–µ–∂–¥—É–Ω–∞—Ä–æ–¥–Ω—ã–µ –æ—Ç–Ω–æ—à–µ–Ω–∏—è', '–ø–æ–ª–∏—Ç–æ–ª–æ–≥–∏—è', '–ø–æ–ª–∏—Ç–æ–ª–æ–≥–∏—è –∏ –º–∏—Ä–æ–≤–∞—è –ø–æ–ª–∏—Ç–∏–∫–∞', '–ø—Ä–æ–≥—Ä–∞–º–º–∞ –¥–≤—É—Ö –¥–∏–ø–ª–æ–º–æ–≤ –Ω–∏—É –≤—à—ç –∏ —É–Ω–∏–≤–µ—Ä—Å–∏—Ç–µ—Ç–∞ –∫—ë–Ω—Ö–∏ ¬´—ç–∫–æ–Ω–æ–º–∏–∫–∞ –∏ –ø–æ–ª–∏—Ç–∏–∫–∞ –≤ –∞–∑–∏–∏¬ª', '–ø—Å–∏—Ö–æ–ª–æ–≥–∏—è', '—Å–æ—Ü–∏–æ–ª–æ–≥–∏—è', '—Å–æ—Ü–∏–æ–ª–æ–≥–∏—è –∏ —Å–æ—Ü–∏–∞–ª—å–Ω–∞—è –∏–Ω—Ñ–æ—Ä–º–∞—Ç–∏–∫–∞', '–≥–ª–æ–±–∞–ª—å–Ω—ã–µ —Ü–∏—Ñ—Ä–æ–≤—ã–µ –∫–æ–º–º—É–Ω–∏–∫–∞—Ü–∏–∏', '–∂—É—Ä–Ω–∞–ª–∏—Å—Ç–∏–∫–∞', '–∫–∏–Ω–æ–ø—Ä–æ–∏–∑–≤–æ–¥—Å—Ç–≤–æ', '–º–µ–¥–∏–∞–∫–æ–º–º—É–Ω–∏–∫–∞—Ü–∏–∏', '—Ä–µ–∫–ª–∞–º–∞ –∏ —Å–≤—è–∑–∏ —Å –æ–±—â–µ—Å—Ç–≤–µ–Ω–Ω–æ—Å—Ç—å—é', '—Å—Ç—Ä–∞—Ç–µ–≥–∏—è –∏ –ø—Ä–æ–¥—é—Å–∏—Ä–æ–≤–∞–Ω–∏–µ –≤ –∫–æ–º–º—É–Ω–∏–∫–∞—Ü–∏—è—Ö', '—É–ø—Ä–∞–≤–ª–µ–Ω–∏–µ –≤ –∫—Ä–µ–∞—Ç–∏–≤–Ω—ã—Ö –∏–Ω–¥—É—Å—Ç—Ä–∏—è—Ö', '–∞–∫—Ç–µ—Ä', '—Ñ–∏–∑–∏–∫–∞', '—Ñ–∏–ª–æ—Å–æ—Ñ–∏—è', '—Ö–∏–º–∏—è', '–∞–Ω–∞–ª–∏—Ç–∏–∫–∞ –≤ —ç–∫–æ–Ω–æ–º–∏–∫–µ', '–±–∏–∑–Ω–µ—Å-–∏–Ω—Ñ–æ—Ä–º–∞—Ç–∏–∫–∞', '–≥–æ—Å—É–¥–∞—Ä—Å—Ç–≤–µ–Ω–Ω–æ–µ –∏ –º—É–Ω–∏—Ü–∏–ø–∞–ª—å–Ω–æ–µ —É–ø—Ä–∞–≤–ª–µ–Ω–∏–µ', '–º–∞—Ä–∫–µ—Ç–∏–Ω–≥ –∏ —Ä—ã–Ω–æ—á–Ω–∞—è –∞–Ω–∞–ª–∏—Ç–∏–∫–∞', '–º–µ–∂–¥—É–Ω–∞—Ä–æ–¥–Ω–∞—è –ø—Ä–æ–≥—Ä–∞–º–º–∞ –ø–æ —ç–∫–æ–Ω–æ–º–∏–∫–µ –∏ —Ñ–∏–Ω–∞–Ω—Å–∞–º', '–º–µ–∂–¥—É–Ω–∞—Ä–æ–¥–Ω—ã–π –±–∞–∫–∞–ª–∞–≤—Ä–∏–∞—Ç –ø–æ –±–∏–∑–Ω–µ—Å—É –∏ —ç–∫–æ–Ω–æ–º–∏–∫–µ', '–º–µ–∂–¥—É–Ω–∞—Ä–æ–¥–Ω—ã–π –±–∏–∑–Ω–µ—Å', '–º–∏—Ä–æ–≤–∞—è —ç–∫–æ–Ω–æ–º–∏–∫–∞', '–ø—Ä–æ–≥—Ä–∞–º–º–∞ –¥–≤—É—Ö –¥–∏–ø–ª–æ–º–æ–≤ –Ω–∏—É –≤—à—ç –∏ —Ä—É—Ç ¬´—ç–∫–æ–Ω–æ–º–∏–∫–∞ –∏ –∏–Ω–∂–µ–Ω–µ—Ä–∏—è —Ç—Ä–∞–Ω—Å–ø–æ—Ä—Ç–Ω—ã—Ö —Å–∏—Å—Ç–µ–º¬ª', '—Å–æ–≤–º–µ—Å—Ç–Ω–∞—è –ø—Ä–æ–≥—Ä–∞–º–º–∞ –ø–æ —ç–∫–æ–Ω–æ–º–∏–∫–µ –Ω–∏—É –≤—à—ç –∏ —Ä—ç—à', '—É–ø—Ä–∞–≤–ª–µ–Ω–∏–µ –±–∏–∑–Ω–µ—Å–æ–º', '—É–ø—Ä–∞–≤–ª–µ–Ω–∏–µ –∏ –∞–Ω–∞–ª–∏—Ç–∏–∫–∞ –≤ –≥–æ—Å—É–¥–∞—Ä—Å—Ç–≤–µ–Ω–Ω–æ–º —Å–µ–∫—Ç–æ—Ä–µ', '—É–ø—Ä–∞–≤–ª–µ–Ω–∏–µ —Ü–µ–ø—è–º–∏ –ø–æ—Å—Ç–∞–≤–æ–∫ –∏ –±–∏–∑–Ω–µ—Å-–∞–Ω–∞–ª–∏—Ç–∏–∫–∞', '—É–ø—Ä–∞–≤–ª–µ–Ω–∏–µ —Ü–∏—Ñ—Ä–æ–≤—ã–º –ø—Ä–æ–¥—É–∫—Ç–æ–º', '—Ü–∏—Ñ—Ä–æ–≤–æ–π –º–∞—Ä–∫–µ—Ç–∏–Ω–≥', '—ç–∫–æ–Ω–æ–º–∏–∫–∞', '—ç–∫–æ–Ω–æ–º–∏–∫–∞ –∏ –∞–Ω–∞–ª–∏–∑ –¥–∞–Ω–Ω—ã—Ö', '—ç–∫–æ–Ω–æ–º–∏–∫–∞ –∏ –±–∏–∑–Ω–µ—Å (–æ—á–Ω–æ-–∑–∞–æ—á–Ω–æ–µ –æ–±—É—á–µ–Ω–∏–µ)', '—ç–∫–æ–Ω–æ–º–∏–∫–∞ –∏ —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞', '—ç–∫–æ–Ω–æ–º–∏—á–µ—Å–∫–∏–π –∞–Ω–∞–ª–∏–∑', '–∏–Ω—Ñ–æ–∫–æ–º–º—É–Ω–∏–∫–∞—Ü–∏–æ–Ω–Ω—ã–µ —Ç–µ—Ö–Ω–æ–ª–æ–≥–∏–∏ –∏ —Å–∏—Å—Ç–µ–º—ã —Å–≤—è–∑–∏', '—é—Ä–∏—Å–ø—Ä—É–¥–µ–Ω—Ü–∏—è', '—é—Ä–∏—Å–ø—Ä—É–¥–µ–Ω—Ü–∏—è: —Ü–∏—Ñ—Ä–æ–≤–æ–π —é—Ä–∏—Å—Ç', '–∏–Ω–æ—Å—Ç—Ä–∞–Ω–Ω—ã–µ —è–∑—ã–∫–∏ –∏ –º–µ–∂–∫—É–ª—å—Ç—É—Ä–Ω–∞—è –±–∏–∑–Ω–µ—Å-–∫–æ–º–º—É–Ω–∏–∫–∞—Ü–∏—è', '–∏–Ω–æ—Å—Ç—Ä–∞–Ω–Ω—ã–µ —è–∑—ã–∫–∏ –∏ –º–µ–∂–∫—É–ª—å—Ç—É—Ä–Ω–∞—è –∫–æ–º–º—É–Ω–∏–∫–∞—Ü–∏—è', '–∏–Ω–æ—Å—Ç—Ä–∞–Ω–Ω—ã–µ —è–∑—ã–∫–∏ –∏ –º–µ–∂–∫—É–ª—å—Ç—É—Ä–Ω–∞—è –∫–æ–º–º—É–Ω–∏–∫–∞—Ü–∏—è –≤ –±–∏–∑–Ω–µ—Å–µ', '—Ñ–∏–ª–æ–ª–æ–≥–∏—è', '—Ñ—É–Ω–¥–∞–º–µ–Ω—Ç–∞–ª—å–Ω–∞—è –∏ –∫–æ–º–ø—å—é—Ç–µ—Ä–Ω–∞—è –ª–∏–Ω–≥–≤–∏—Å—Ç–∏–∫–∞', '—Ñ—É–Ω–¥–∞–º–µ–Ω—Ç–∞–ª—å–Ω–∞—è –∏ –ø—Ä–∏–∫–ª–∞–¥–Ω–∞—è –ª–∏–Ω–≥–≤–∏—Å—Ç–∏–∫–∞']
YN = ['–¥–∞', '–Ω–µ—Ç']
EXAM_QUESTIONS = {
    '–ö–∞–∫ –Ω–∞–∑—ã–≤–∞–µ—Ç—Å—è —Å–æ–∑–≤—É—á–∏–µ –∫–æ–Ω—Ü–æ–≤ —Å—Ç—Ä–æ–∫ –≤ —Å—Ç–∏—Ö–æ—Ç–≤–æ—Ä–µ–Ω–∏–∏ (–æ–ø—É—Å—Ç–µ–ª–æ–º ‚Äî —Ç–µ–ª–æ–º; —Ç–∏—à–∏–Ω—ã ‚Äî —Å—Ç–µ–Ω—ã)?': '—Ä–∏—Ñ–º–∞',
    '–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ –º–æ—Ç–æ—Ä —Ö–æ–ª–æ–¥–∏–ª—å–Ω–∏–∫–∞ –ø—Ä–æ—Å–ª—É–∂–∏—Ç –±–æ–ª–µ–µ 1 –≥–æ–¥–∞, —Ä–∞–≤–Ω–∞ 0,8, –∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ –æ–Ω –ø—Ä–æ—Å–ª—É–∂–∏—Ç –±–æ–ª–µ–µ 2 –ª–µ—Ç, —Ä–∞–≤–Ω–∞ 0,6. –ö–∞–∫–æ–≤–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ –º–æ—Ç–æ—Ä –ø—Ä–æ—Å–ª—É–∂–∏—Ç –±–æ–ª–µ–µ 1 –≥–æ–¥–∞, –Ω–æ –Ω–µ –±–æ–ª–µ–µ 2 –ª–µ—Ç?': '0.2',
    '–†–∞—Å—Å—Ç–∞–≤—å—Ç–µ –∑–Ω–∞–∫–∏ –ø—Ä–µ–ø–∏–Ω–∞–Ω–∏—è: —É–∫–∞–∂–∏—Ç–µ –≤—Å–µ —Ü–∏—Ñ—Ä—ã, –Ω–∞ –º–µ—Å—Ç–µ –∫–æ—Ç–æ—Ä—ã—Ö –≤ –ø—Ä–µ–¥–ª–æ–∂–µ–Ω–∏–∏ –¥–æ–ª–∂–Ω—ã —Å—Ç–æ—è—Ç—å –∑–∞–ø—è—Ç—ã–µ.\n–ò–ª—é—à–∞ –±—ã–ª –≤–ø–µ—á–∞—Ç–ª–∏—Ç–µ–ª—å–Ω—ã–º —Ä–µ–±—ë–Ω–∫–æ–º (1) –∏ –æ—á–µ–Ω—å –ª—é–±–∏–ª —Å–ª—É—à–∞—Ç—å —Å–∫–∞–∑–∫–∏ (2) –∫–æ—Ç–æ—Ä—ã–µ —Ä–∞—Å—Å–∫–∞–∑—ã–≤–∞–ª–∞ –µ–º—É –Ω—è–Ω—è (3) –ø–æ—Ç–æ–º—É —á—Ç–æ –≤ –Ω–∏—Ö –±—ã–ª–æ –º–Ω–æ–≥–æ —á—É–¥–µ—Å–Ω–æ–≥–æ (4) –∞ (5) –¥–æ–±—Ä–æ –≤—Å–µ–≥–¥–∞ –ø–æ–±–µ–∂–¥–∞–ª–æ –∑–ª–æ.': '234'
}
COMPANIES = ['–û–û–û "–Ø–ù–î–ï–ö–°"', '–û–û–û "–í –ö–û–ù–¢–ê–ö–¢–ï"', '–û–û–û "–ê–í–ò–¢–û"', '–û–û–û "–í–´–°–®–ê–Ø –®–ö–û–õ–ê –°–ï–†–§ –ö–û–§–ï –•–ò–õ–õ–°"']


# vars
state = {
    'user_name': None,
    'program': None,
    'is_commercial': None,
    'primary_way': None,
    'is_accepted': False,
}
required_documents = {
    '–°–ù–ò–õ–°': '.jpg', 
    '–ü–∞—Å–ø–æ—Ä—Ç (1 —Å—Ç—Ä.)': '.png',
    '–ü–∞—Å–ø–æ—Ä—Ç (—Å—Ç—Ä. —Å –ø—Ä–æ–ø–∏—Å–∫–æ–π)': '.webp',
    '–ê—Ç—Ç–µ—Å—Ç–∞—Ç': '.pdf',
    '–°–æ–≥–ª–∞—Å–∏–µ –æ–± –æ–±—Ä–∞–±–æ—Ç–∫–µ –ü–î': '.docx',
    '–§–æ—Ç–æ 3 –Ω–∞ 4': '.tiff',
    '–†–∞–∑–º–µ—Ä –æ–±—É–≤–∏': '.txt',
    '–î–∏–ø–ª–æ–º "–†—É—Å—Å–∫–∏–π –ú–µ–¥–≤–µ–∂–æ–Ω–æ–∫"': '.pdf',
    '–§–æ—Ç–æ –∫—Ä–µ–¥–∏—Ç–Ω–æ–π –∫–∞—Ä—Ç—ã —Å –¥–≤—É—Ö —Å—Ç–æ—Ä–æ–Ω': '.heif',
}
application_troubles = ['file_size', 'file_name', 'server_error', 'server_maintainance', 'server_flood', 'server_hacked']


# main code
print('–ø—Ä–∏–≤–µ—Ç! –¥–æ–±—Ä–æ –ø–æ–∂–∞–ª–æ–≤–∞—Ç—å –≤ —Å–∏–º—É–ª—è—Ç–æ—Ä –ø–æ—Å—Ç—É–ø–ª–µ–Ω–∏—è –≤ –Ω–∏—É –≤—à—ç.\n—ç—Ç–æ –æ—á–µ–Ω—å —Å–ª–æ–∂–Ω—ã–π –∏ –Ω–µ–ø–æ–Ω—è—Ç–Ω—ã–π –ø—É—Ç—å, –∏ –∏–Ω—Ç–µ—Ä–µ—Å–Ω–æ –Ω–µ –±—É–¥–µ—Ç.\n\n–¥–ª—è –Ω–∞—á–∞–ª–∞ ‚Äî –¥–∞–≤–∞–π –∑–Ω–∞–∫–æ–º–∏—Ç—å—Å—è.')
state['user_name'] = ask('–∫–∞–∫ —Ç–µ–±—è –∑–æ–≤—É—Ç?', lambda x: (len(x) > 1))

rnd = Random(state['user_name'])

space()
print(f"{state['user_name'].capitalize()}, –Ω–∞–º –Ω–∞–¥–æ –≤—ã–±—Ä–∞—Ç—å –û–ü. —ç—Ç–æ —Ç–∞–∫–∞—è —à—Ç—É–∫–∞, –Ω–∞ –∫–æ—Ç–æ—Ä—É—é –ø—Ä–∏–¥–µ—Ç—Å—è –ø–æ—Ç—Ä–∞—Ç–∏—Ç—å –±–ª–∏–∂–∞–π—à–∏–µ 4 –≥–æ–¥–∞.\n–≤—ã–±—Ä–∞—Ç—å –ø–æ–¥—Ö–æ–¥—è—â—É—é –º–æ–∂–Ω–æ –Ω–∞ —Å–∞–π—Ç–µ –≤—ã—à–∫–∏: https://www.hse.ru/education/\n\n–ø–æ–∂–∞–ª—É–π—Å—Ç–∞, –Ω–∞–ø–∏—à–∏, –∫–∞–∫—É—é –û–ü —Ç—ã –≤—ã–±—Ä–∞–ª(–∞).")
state['program'] = ask('–Ω–∞ –∫–∞–∫—É—é –æ–ø –ø–æ—Å—Ç—É–ø–∞–µ–º? –Ω–∞–ø—Ä–∏–º–µ—Ä, "–¥–∏–∑–∞–π–Ω"', lambda x: x.lower() in PROGRAMS)

space()
print(f'–∏–Ω—Ç–µ—Ä–µ—Å–Ω—ã–π –≤—ã–±–æ—Ä. –≥–æ–¥ –æ–±—É—á–µ–Ω–∏—è –Ω–∞ —ç—Ç–æ–π –ø—Ä–æ–≥—Ä–∞–º–º–µ —Å—Ç–æ–∏—Ç {rnd.get(0.1) + 1 * 345678} —Ä—É–±–ª–µ–π.\n–µ—Å—Ç—å –ª–∏ —É —Ç–µ–±—è —Ç–∞–∫–∏–µ –¥–µ–Ω—å–≥–∏?')
state['is_commercial'] = ask_yn()

space()
if state['is_commercial']:
    exam_required_score = 180
    print('–≥—É–¥—Ñ–æ—é! —Ç–æ–≥–¥–∞ –¥–ª—è –ø–æ—Å—Ç—É–ø–ª–µ–Ω–∏—è –Ω—É–∂–Ω–æ –Ω–∞–±—Ä–∞—Ç—å –≤—Å–µ–≥–æ 180 –±–∞–ª–ª–æ–≤ –µ–≥—ç –≤ —Å—É–º–º–µ.')
    state['primary_way'] = 'exam'
else:
    exam_required_score = 300 - rnd.get(0.3)
    print(f'–Ω–µ—É–¥–∏–≤–∏—Ç–µ–ª—å–Ω–æ. —Ç–æ–≥–¥–∞ –ø—Ä–∏–¥–µ—Ç—Å—è –ø–æ—Å—Ç—É–ø–∞—Ç—å –±–µ—Å–ø–ª–∞—Ç–Ω–æ. –¥–∞–≤–∞–π –æ—Ü–µ–Ω–∏–º —Ç–≤–æ–∏ —à–∞–Ω—Å—ã –Ω–∞ —ç—Ç–æ.\n\n–ø–æ –µ–≥—ç –¥–ª—è –±—é–¥–∂–µ—Ç–∞ –ø—Ä–∏–¥–µ—Ç—Å—è –Ω–∞–±—Ä–∞—Ç—å {exam_required_score} –±–∞–ª–ª–æ–≤. –û—Å–∏–ª–∏–º?')
    if ask_yn():
        space()
        print('—Ö–æ—Ä–æ—à–æ, –±—É–¥–µ–º –ø—ã—Ç–∞—Ç—å—Å—è –ø–æ—Å—Ç—É–ø–∞—Ç—å –ø–æ –µ–≥—ç.')
        state['primary_way'] = 'exam'
    else:
        space()
        print('–Ω—É –∏ –ø—Ä–∞–≤–∏–ª—å–Ω–æ, –ª—É—á—à–µ –¥–∞–∂–µ –Ω–µ –ø—ã—Ç–∞—Ç—å—Å—è.\n—Ç–æ–≥–¥–∞ –º–æ–∂–µ—Ç —Å–æ—Ü–∏–∞–ª—å–Ω—ã–π –ª–∏—Ñ—Ç ‚Äî —Ç—ã —Å–ª—É—á–∞–π–Ω–æ –Ω–µ –∏–∑ –Ω–µ–±–ª–∞–≥–æ–ø–æ–ª—É—á–Ω–æ–π —Å–µ–º—å–∏?')
        if ask_yn():
            exam_required_score = int(exam_required_score * 0.75)
            space()
            print(f'—ç—Ç–æ –Ω–µ–ø–ª–æ—Ö–æ –ø–æ–≤—ã—à–∞–µ—Ç —Ç–≤–æ–∏ —à–∞–Ω—Å—ã! –±—É–¥–µ—à—å –≤ –æ—Ç–¥–µ–ª—å–Ω–æ–º —Å–ø–∏—Å–∫–µ –Ω–∞ –ø–æ—Å—Ç—É–ø–ª–µ–Ω–∏–µ, –≥–¥–µ –ø—Ä–æ—Ö–æ–¥–Ω–æ–π –±–∞–ª–ª ‚Äî {exam_required_score}. —É–∂–µ –ª—É—á—à–µ!')
            state['primary_way'] = 'exam'
        else:
            space()
            print('–≤ –¥–∞–Ω–Ω–æ–º –∫–æ–Ω—Ç–µ–∫—Å—Ç–µ –¥–∞–∂–µ —Ö–æ—á–µ—Ç—Å—è —Å–∫–∞–∑–∞—Ç—å "–∂–∞–ª—å". –Ω—É –ª–∞–¥–Ω–æ.\n—Å–ª–µ–¥—É—é—â–∏–π –≤–∞—Ä–∏–∞–Ω—Ç ‚Äî —Ü–µ–ª–µ–≤–∞—è –∫–≤–æ—Ç–∞. –ø–æ–ø—Ä–æ–±—É–µ–º –Ω–∞–π—Ç–∏ –∫–æ–º–ø–∞–Ω–∏—é, –∫–æ—Ç–æ—Ä–∞—è –∑–∞—Ö–æ—á–µ—Ç –æ–ø–ª–∞—Ç–∏—Ç—å –æ–±—É—á–µ–Ω–∏–µ –∑–∞ —Ç–µ–±—è?')
            if ask_yn():
                space()
                print(f'–∑–¥–æ—Ä–æ–≤–æ! –¥–∞–≤–∞–π –ø—Ä–æ–±–æ–≤–∞—Ç—å')
                state['primary_way'] = 'quota'
            else:
                space()
                print('—á—Ç–æ-–∂, –Ω–∞—à–∞ –ø–æ—Å–ª–µ–¥–Ω—è—è –æ–ø—Ü–∏—è ‚Äî –æ–ª–∏–º–ø–∏–∞–¥—ã. –¥—Ä—É–≥–∏—Ö –≤–∞—Ä–∏–∞–Ω—Ç–æ–≤ –Ω–µ—Ç.')
                state['primary_way'] = 'olymp'                 

space()
print('–ø—Ä–∏—Å—Ç—É–ø–∞–µ–º –∫ –Ω–µ–ø–æ—Å—Ä–µ–¥—Å—Ç–≤–µ–Ω–Ω–æ–º—É –ø–æ—Å—Ç—É–ø–ª–µ–Ω–∏—é. —ç—Ç–æ —Å–∏–º—É–ª—è—Ç–æ—Ä, –ø–æ—ç—Ç–æ–º—É –µ—Å–ª–∏ –Ω–µ –≤—ã–π–¥–µ—Ç ‚Äî –≤—Å–µ–≥–¥–∞ –º–æ–∂–Ω–æ –ø–æ–ø—Ä–æ–±–æ–≤–∞—Ç—å —Å–Ω–æ–≤–∞. –ø–æ–∫–∞ –Ω–µ –Ω–∞–¥–æ–µ—Å—Ç. –Ω—É, –∏–ª–∏ –ø–æ–∫–∞ –Ω–µ –ø–æ—Å—Ç—É–ø–∏—à—å. –≥–æ—Ç–æ–≤?')
ask_yn_force()

while True:
    space()
    if state['primary_way'] == 'exam':
        print('—Å–¥–∞–µ–º –µ–≥—ç. –Ω–µ –Ω–µ—Ä–≤–Ω–∏—á–∞–µ–º, –≤—Å–ø–æ–º–∏–Ω–∞–µ–º –∫ —á–µ–º—É –≥–æ—Ç–æ–≤–∏–ª–∏—Å—å.')
        score = 0
        for k, v in EXAM_QUESTIONS.items():
            space()
            print(k)
            ans = ask('—Ç–≤–æ–π –æ—Ç–≤–µ—Ç')
            if ans == v:
                score += 100
        
        space()
        score += rnd.get(0.4)

        if score >= exam_required_score:
            print(f'—Ç—ã –Ω–∞–±—Ä–∞–ª {score} –±–∞–ª–ª–æ–≤ –ø—Ä–∏ –ø–æ—Ä–æ–≥–µ –≤ {exam_required_score}. —É—Ä–∞!!! –ø–æ–∑–¥—Ä–∞–≤–ª—è—é!!!')
            state['is_accepted'] = True
        else:
            print(f'—Ç—ã –Ω–∞–±—Ä–∞–ª {score} –±–∞–ª–ª–æ–≤, –∞ –Ω–∞–¥–æ –±—ã–ª–æ {exam_required_score}. —É–≤—ã... –ø–æ—Å—Ç—É–ø–ª–µ–Ω–∏–µ –Ω–∞ —ç—Ç–æ–º –∑–∞–∫–æ–Ω—á–∏–ª–æ—Å—å.')
            state['is_accepted'] = False

    elif state['primary_way'] == 'quota':
        print('–µ—Å—Ç—å –Ω–µ—Å–∫–æ–ª—å–∫–æ –∫–æ–º–ø–∞–Ω–∏–π. –∑–∞–∏–Ω—Ç–µ—Ä–µ—Å—É–µ—à—å –ª–∏ —Ç—ã –∏—Ö ‚Äî –≤–æ–ø—Ä–æ—Å. –Ω–æ –≤—ã–±—Ä–∞—Ç—å –º–æ–∂–Ω–æ —Ç–æ–ª—å–∫–æ –æ–¥–Ω—É. –≤—ã–±–∏—Ä–∞–π!')
        for index, i in enumerate(COMPANIES):
            print(f'{index + 1}. {i}')
        choice = ask('–≤–≤–µ–¥–∏ —á–∏—Å–ª–æ', lambda x: x.isdigit() and int(x) > 0 and int(x) <= len(COMPANIES))

        space()
        success = (int(choice) + rnd.get(0.4)) % 2 == 0
        if success:
            print('—ç—Ç–∞ –∫–æ–º–ø–∞–Ω–∏—è –æ–ø–ª–∞—Ç–∏—Ç —Ç–≤–æ–µ –æ–±—É—á–µ–Ω–∏–µ! —É—Ä–∞!!! –ø–æ–∑–¥—Ä–∞–≤–ª—è—é!!!')
            state['is_accepted'] = True
        else:
            print('—ç—Ç–æ–π –∫–æ–º–ø–∞–Ω–∏–∏ –Ω–µ –æ—á–µ–Ω—å –∏–Ω—Ç–µ—Ä–µ—Å–Ω–æ. —É–≤—ã... –ø–æ—Å—Ç—É–ø–ª–µ–Ω–∏–µ –Ω–∞ —ç—Ç–æ–º –∑–∞–∫–æ–Ω—á–∏–ª–æ—Å—å.')
            state['is_accepted'] = False

    elif state['primary_way'] == 'olymp' or True:
        print('–µ–≥—ç, –∫–∞–∫ –∏–∑–≤–µ—Å—Ç–Ω–æ, –ª–æ—Ç–µ—Ä–µ—è, –Ω–æ –æ–ª–∏–º–ø–∏–∞–¥—ã ‚Äî –ª–æ—Ç–µ—Ä–µ—è –≤ –∫—É–±–µ. –ø–æ—ç—Ç–æ–º—É —Ç—É—Ç –≤ —Ü–µ–ª—è—Ö —Å–∏–º—É–ª—è—Ü–∏–∏ –¥–∞–∂–µ –Ω–µ –±—É–¥–µ—Ç —Ä–µ–∞–ª—å–Ω—ã—Ö –≤–æ–ø—Ä–æ—Å–æ–≤ ‚Äî —è –ø—Ä–µ–¥–ª–∞–≥–∞—é —Ç–∫–Ω—É—Ç—å –ø–∞–ª—å—Ü–µ–º –≤ –Ω–µ–±–æ –∏ –Ω–∞–∑–≤–∞—Ç—å —á–∏—Å–ª–æ –æ—Ç 1 –¥–æ 256.')
        choice = ask('–≤–≤–µ–¥–∏ —á–∏—Å–ª–æ', lambda x: x.isdigit() and int(x) > 0 and int(x) <= 256)
        
        space()
        success = (int(choice) + rnd.get(0.4)) % 3 == 0
        if success:
            print('—Ç—ã –≤ —á–∏—Å–ª–µ –ø–æ–±–µ–¥–∏—Ç–µ–ª–µ–π! —É—Ä–∞!!! –ø–æ–∑–¥—Ä–∞–≤–ª—è—é!!!')
            state['is_accepted'] = True
        else:
            print(f'—É–ø—Å, —Ç–µ–±—è –æ–±–æ—à–ª–æ {125 * (rnd.get(0.5) + int(choice))} –¥—Ä—É–≥–∏—Ö —É—á–µ–Ω–∏–∫–æ–≤. —É–≤—ã... –ø–æ—Å—Ç—É–ø–ª–µ–Ω–∏–µ –Ω–∞ —ç—Ç–æ–º –∑–∞–∫–æ–Ω—á–∏–ª–æ—Å—å.')
            state['is_accepted'] = False


    if state['is_accepted']:
        break
    else:
        space()
        print('–ø—Ä–æ–±—É–µ–º —Å–Ω–æ–≤–∞?')
        if not ask_yn(): break

space()
print('–ø—Ä–∏—à–ª–∞ –ø–æ—Ä–∞ –ø–æ–¥–∞–≤–∞—Ç—å –¥–æ–∫—É–º–µ–Ω—Ç—ã. –≤—Å–µ, —á—Ç–æ –±—ã–ª–æ –¥–æ —ç—Ç–æ–≥–æ, –±—ã–ª–æ –≥–æ—Ä–∞–∑–¥–æ –ª–µ–≥—á–µ, —Ç–∞–∫ —á—Ç–æ —É–¥–∞—á–∏. –µ—Å–ª–∏ —á—Ç–æ-—Ç–æ –ø–æ–π–¥–µ—Ç –Ω–µ —Ç–∞–∫ ‚Äî –ø—Ä–∏–¥–µ—Ç—Å—è –Ω–∞—á–∏–Ω–∞—Ç—å —Å–Ω–∞—á–∞–ª–∞.\n–≤–æ—Ç —Å–ø–∏—Å–æ–∫ –¥–æ–∫—É–º–µ–Ω—Ç–æ–≤:')

space()
for k, v in required_documents.items():
    print(f'{k} ‚Äî —Å–ª–∞—Ç—å –≤ —Ñ–æ—Ä–º–∞—Ç–µ {v}')

space()
print('–ø—Ä–∏—Å—Ç—É–ø–∞–µ–º?')
ask_yn_force()

attempt = 0
while True:
    for k, v in required_documents.items():
        space()
        print(f'–ø—Ä–∏–ª–æ–∂–∏ {k}')
        file = ask('–Ω–∞–ø–∏—à–∏ –Ω–∞–∑–≤–∞–Ω–∏–µ —Ñ–∞–π–ª–∞ –≤ –Ω—É–∂–Ω–æ–º —Ñ–æ—Ä–º–∞—Ç–µ')
        attempt += 1
        if not file.endswith(v):
            print(f'–Ω–µ–≤–µ—Ä–Ω—ã–π —Ñ–æ—Ä–º–∞—Ç —Ñ–∞–π–ª–∞! –Ω—É–∂–µ–Ω –±—ã–ª {v}. –∑–∞–Ω–æ–≤–æ.')
            break
        if attempt % 4 == 0 and len(application_troubles) > 0:
            trouble = application_troubles.pop(0)
            if trouble == 'file_size':
                print('–æ–π! —Ä–∞–∑–º–µ—Ä —Å–ª–∏—à–∫–æ–º –±–æ–ª—å—à–æ–π. –∑–∞–Ω–æ–≤–æ.')
            elif trouble == 'file_name':
                print('–æ–π! –∏–º—è –Ω–µ –ø–æ–¥—Ö–æ–¥–∏—Ç. –∑–∞–Ω–æ–≤–æ.')
            elif trouble == 'server_error':
                print('–æ–π! —Å–µ—Ä–≤–µ—Ä–Ω–∞—è –æ—à–∏–±–∫–∞. –∑–∞–Ω–æ–≤–æ.')
            elif trouble == 'server_maintainance':
                print('–æ–π! –Ω–∞ —Å–µ—Ä–≤–µ—Ä–µ —Ä–∞–±–æ—Ç—ã. –∑–∞–Ω–æ–≤–æ.')
            elif trouble == 'server_flood':
                print('–æ–π! —Å–µ—Ä–≤–µ—Ä –∑–∞—Ç–æ–ø–∏–ª–æ... –±–ª–±–ª–±–ª–±–ª–ª...üêüüêüüêü... –∑–∞–Ω–æ–≤–æ.')
            elif trouble == 'server_hacked':
                print('–æ–π! —Å–µ—Ä–≤–µ—Ä –≤–∑–ª–æ–º–∞–ª–∏... Ëµ∑‰æÜÔºÅ‰∏çÈ°òÂÅöÂ•¥Èö∏ÁöÑ‰∫∫ÂÄëÔºÅ–∑–∞–Ω–æ–≤–æ.')
            else:
                print('–æ–π! –Ω–µ–∏–∑–≤–µ—Å—Ç–Ω–∞—è –æ—à–∏–±–∫–∞. –∑–∞–Ω–æ–≤–æ.')
            break
    else:
        break

space(2)
print('—É—Ä–∞! —Ç–≤–æ—è –∑–∞—è–≤–∫–∞ –≥–æ—Ç–æ–≤–∞. –æ—Ç–ø—Ä–∞–≤–ª—è–µ–º??')
ask_yn_force()

space()
print('500 Internal Server Error')

story = '—Ç—ã –ø—ã—Ç–∞–ª—Å—è –ø–æ—Å—Ç—É–ø–∏—Ç—å –≤ –Ω–∏—É –≤—à—ç.\n'

for k, v in state.items():
    if k == 'user_name':
        story += f'—Ç–µ–±—è –∑–≤–∞–ª–∏ {v}\n'
    elif k == 'program':
        story += f'—Ç—ã —Ö–æ—Ç–µ–ª–∏ –Ω–∞ –æ–ø {v}\n'
    elif k == 'is_commercial':
        story += f"–≤–∞—Ä–∏–∞–Ω—Ç ‚Äî {'–∫–æ–º–º–µ—Ä—Ü–∏—è' if v else '–±—é–¥–∂–µ—Ç'}\n"
    elif k == 'is_accepted':
        story += f"—Ç–µ–±—è {'–≤–∑—è–ª–∏' if v else '–Ω–µ –≤–∑—è–ª–∏'}\n"

with open('story.txt', 'w') as file:
    file.write(story)
```
